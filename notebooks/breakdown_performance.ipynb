# define generalist vs specialist

load data

DATA is a set of experiments, and their performance over three conditions and a number of plume densities. 

We are interested in seeing if there are specialists that emerge out of training, i.e. agents that are particularly good at one condition or densities. 

I will try to identify these agents by highlighting the top 5 agents, eval'd on full plume in various conditions. See if there will be un-highlighted agents that are better than the highlighted ones. 

```python
file_path = '/src/TrainigCurriculum/JH_boilerplate/performance_eval/performance_all_Expt.tsv'
now_delimiter = '\t'
# read the text file into a pandas dataframe
df = pd.read_csv(file_path, delimiter=now_delimiter)
df['condition'] = df['condition'].astype('category')
df['seed'] = df['seed'].astype('category')

df['experiment'] = [ps.split('/')[1] for ps in df['model_dir']]
```

```python
df['model_dir'] = ['/src/TrainigCurriculum/data/published_results/reproduce' +p for p in df['model_dir']]
```

```python
df_pivot = df.pivot(index=['model_dir', 'experiment'], columns='dataset', values='HOME').reset_index()
df_pivot['total'] = df_pivot[['constantx5b5', 'noisy3x5b5', 'switch45x5b5',]].sum(axis=1, skipna=True)
col_order = [
        'total',
        'constantx5b5', 
        'switch45x5b5', 
        'noisy3x5b5', 
#         'noisy6x5b5',
#         'switch30x5b5', 
#         'switch15x5b5', 
        'constantx5b5_0.8',
        'constantx5b5_0.6',
        'constantx5b5_0.4', 
        'constantx5b5_0.2',
        'experiment',
        'model_dir',
]
df_pivot = df_pivot[col_order]
df_pivot = df_pivot.sort_values(by='total', ascending=False)
df_pivot.head()

```

```python
# tmp = df_pivot.melt(id_vars=['model_dir', 'experiment', 'total'], var_name='dataset', value_name='HOME')
tmp = df_pivot[df_pivot.experiment == 'ExptMemory20230922']

COI = ['constantx5b5', 'switch45x5b5', 'noisy3x5b5']
for c in COI:
    
    # Find the top 5 rows
    top_rows = tmp.nlargest(5, c)

    # Create a new column 'is_top' that is True for the top rows and False otherwise
    top_at = c + '_top'
    tmp[top_at] = False
    tmp.loc[tmp.index.isin(top_rows.index), top_at] = True
    # tmp.loc[top_at][tmp.index.isin(top_rows.index)] = True
tmp = tmp.sort_values(by='total', ascending=False)
tmp = tmp.reset_index(drop=True)
tmp.index = tmp.index + 1
col_order = [
'top', 'total', 'constantx5b5', 'switch45x5b5', 'noisy3x5b5',
       'constantx5b5_0.8', 'constantx5b5_0.6', 'constantx5b5_0.4',
       'constantx5b5_0.2', 'experiment', 'model_dir', 'constantx5b5_top',
       'noisy3x5b5_top', 'switch45x5b5_top', 
]
tmp['top'] = tmp[['constantx5b5_top', 'noisy3x5b5_top', 'switch45x5b5_top']].sum(axis=1, skipna=True)



tmp = tmp[col_order]
tmp
```
