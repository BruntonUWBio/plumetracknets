```python
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html
from scipy.stats import poisson
mu = 0.9
mean, var, skew, kurt = poisson.stats(mu, moments='mvsk')
fig, ax = plt.subplots(1, 1,figsize=(3,1.5))
# fig, ax = plt.subplots(1, 1)
# fig, ax = plt.subplots(1, 1,figsize=(0.62,0.3))
x = np.arange(poisson.ppf(0.01, mu), poisson.ppf(0.9999, mu))
ax.plot(x, poisson.pmf(x, mu), 'bo', ms=7, label='poisson pmf')
ax.vlines(x, 0, poisson.pmf(x, mu), colors='b', lw=4, alpha=0.6)
plt.xticks([])
plt.yticks([])
# plt.tight_layout()

```
